```python
##################################### Genetic Algorithm Demo #############################################
import matplotlib.pyplot as plt 
import numpy
import ga_min   # import GA.py module
%matplotlib inline

"""
The y=target is to maximize/minimize this equation ASAP:
    y = w1x1+w2x2+w3x3+w4x4+w5x5
    where (x1,x2,x3,x4,x5,x6)=(4,-2,3.5,5,-11)
    What are the best values for the 6 weights w1 to w5?
    We are going to use the genetic algorithm for the best possible values after a number of generations.
    OR,
    a + 2b + 3c + 4d = 30
    y = a + 2b + 3c + 4d - 30
    a = ? b = ? c = ? d = ?
"""

# Inputs of the equation.
equation_inputs = [1,2,3,4]

# Number of the weights we are looking to optimize.
num_weights = len(equation_inputs)

"""
Genetic algorithm parameters:
    Mating pool size
    Population size
"""
sol_per_pop = 10
cr=0.8
num_parents_mating = round(sol_per_pop*cr)

# Defining the population size.
# The population will have sol_per_pop chromosome where each chromosome has num_weights genes.
pop_size = (sol_per_pop,num_weights) 
#Creating the initial population.
new_population = numpy.random.uniform(low=0, high=30, size=pop_size)   # 0 to 30
print('Initial Population:\n',new_population,'\n')

fig, ax = plt.subplots()

num_generations = 1000   # Max Iteration
for generation in range(num_generations):
    # print("\nGeneration: ", generation)
    # Measing the fitness of each chromosome in the population.
    fitness = ga_min.cal_pop_fitness(equation_inputs, new_population)

    # Selecting the best parents in the population for mating.
    parents = ga_min.select_mating_pool(new_population, fitness, 
                                      num_parents_mating)

    # Generating next generation using crossover.
    offspring_crossover = ga_min.crossover(parents,
                                       offspring_size=(pop_size[0]-parents.shape[0], num_weights))

    # Adding some variations to the offsrping using mutation.
    offspring_mutation = ga_min.mutation(offspring_crossover)

    # Creating the new population based on the parents and offspring.
    new_population[0:parents.shape[0], :] = parents
    new_population[parents.shape[0]:, :] = offspring_mutation
    new_population = numpy.absolute(new_population)
    
    # The best result in the current iteration.
    print("Best result: ", numpy.min(numpy.sum(new_population*equation_inputs, axis=1)))

    ax.scatter(new_population[:, 0], new_population[:, 1], c = 'blue', s = 12)
       
ax.set_xlim((0, 30)) 
ax.set_ylim((0, 30)) 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("{} vs {}")
# set a title and labels
ax.set_title('Search Space')

# Getting the best solution after iterating finishing all generations.
#At first, the fitness is calculated for each solution in the final generation.
fitness = ga_min.cal_pop_fitness(equation_inputs, new_population)
# Then return the index of that solution corresponding to the best fitness.
best_match_idx = numpy.where(fitness == numpy.min(fitness))   #min

print("\nLast Population:\n")
print(new_population)
print("\nLast Fitness:\n")
print(fitness)
print("\nBest solution: ", new_population[best_match_idx, :])
print("Best solution fitness: ", fitness[best_match_idx])
```
